```python
# Import dependencies. The pyplot module will allow us to create all the necessary charts; 
# create a plotting area in a figure; plot lines in a plotting area; and annotate the chart with labels, 
# colors, and other decorative features.

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
#change the colors of the markers to red
plt.scatter(x_axis, y_axis, c="red")
#add a legend for the city of Chicago
plt.scatter(x_axis, y_axis, c="red", label = "Chicago")
plt.legend()
#add a title and axes label
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
```

```python
#switch the axis s that the Fare($) data is on the x-axis
plt.scatter(y_axis, x_axis, c='red',label = "Chicago")
plt.legend()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
```

```python
plt.scatter(y_axis, x_axis, c='red',label = "Chicago")
plt.legend()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
#add a limit for the x-axis data
plt.xlim(5,45)
#invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#make dots in y-axis larger with for loop to increase value
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#make dots larger by using python's list comprehension to replace for loop
# syntax: new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#create a scatter plot using objct-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create a bubble chart using objct-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#using object-oriented approach change color to skyblue
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis, c="skyblue")
```

```python
#change size of the marker to 5 times each data point
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis], c="skyblue")
```

```python
#change the marker color to 20% transparent
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis], c="skyblue", alpha=0.8)
```

```python
#add black edge color to marker, make linewidth 2 points, add legend for city of Boston
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis], c="skyblue", alpha=0.8, edgecolor="black", linewidth=2, label="Boston")
#add a legend for the city of Boston
ax.legend()
#add a title
ax.set_title("PyBer Fare by Month")
```

```python
#Create a piechart
plt.pie(y_axis, labels=x_axis)
```

```python
#increase figure size
plt.subplots(figsize=(8,8))
#format piechart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
```
